```python
d2l.DATA_HUB['pokemon'] = (d2l.DATA_URL + 'pokemon.zip',
                           'c065c0e2593b8b161a2d7873e42418bf6a21106c')

data_dir = d2l.download_extract('pokemon')
```

```python
device = d2l.try_gpu()
```

```python
pokemon = torchvision.datasets.ImageFolder(data_dir)
```

```python
batch_size = 21
transformer = torchvision.transforms.Compose([
    # reducing HxW in order to simplify the rpoblem, thus improving performance
    torchvision.transforms.Resize((32, 32)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(0.5, 0.5)
])
pokemon.transform = transformer
data_iter = torch.utils.data.DataLoader(
    pokemon, batch_size=batch_size,
    shuffle=True, num_workers=d2l.get_dataloader_workers())
```

```python
warnings.filterwarnings('ignore')
d2l.set_figsize((16, 16))
for X, y in data_iter:
    imgs = X[0:20,:,:,:].permute(0, 2, 3, 1)/2+0.5
    d2l.show_images(imgs, num_rows=4, num_cols=5)
    break
```

```python
class GeneratorBlock(nn.Module):
    def __init__(self, out_channels, in_channels=3, kernel_size=4, strides=2,
                 padding=1, **kwargs):
        super(GeneratorBlock, self).__init__(**kwargs)
        self.conv2d_trans = nn.ConvTranspose2d(in_channels, out_channels,
                                kernel_size, strides, padding, bias=False)
        self.batch_norm = nn.BatchNorm2d(out_channels)
        self.activation = nn.GELU()

    def forward(self, X):
        return self.activation(self.batch_norm(self.conv2d_trans(X)))
```

```python
class DiscriminatorBlock(nn.Module):
    def __init__(self, out_channels, in_channels=3, kernel_size=4, strides=2,
                padding=1, alpha=0.2, **kwargs):
        super(DiscriminatorBlock, self).__init__(**kwargs)
        self.conv2d = nn.Conv2d(in_channels, out_channels, kernel_size,
                                strides, padding, bias=False)
        self.batch_norm = nn.BatchNorm2d(out_channels)
        self.activation = nn.GELU()

    def forward(self, X):
        return self.activation(self.batch_norm(self.conv2d(X)))
```

```python
def train(net_Disc, net_Gen, data_iter, num_epochs, lr, latent_dim, device=d2l.try_gpu(), loss_fn=nn.BCEWithLogitsLoss(reduction='sum')):
    loss = loss_fn
    
    for w in net_Disc.parameters():
        nn.init.normal_(w, 0, 0.02)
    for w in net_Gen.parameters():
        nn.init.normal_(w, 0, 0.02)
    
    net_Disc, net_Gen = net_Disc.to(device), net_Gen.to(device)
    
    trainer_hyp_par = {'lr': lr, 'betas': [0.5, 0.999], 'weight_decay': 10e-4}
    
    trainer_Disc = torch.optim.Adam(net_Disc.parameters(), **trainer_hyp_par)
    trainer_Gen  = torch.optim.Adam(net_Gen.parameters(),  **trainer_hyp_par)
    
    animator = d2l.Animator(xlabel='epoch', ylabel='loss', xlim=[1, num_epochs], nrows=2, 
                            figsize=(16, 16), legend=['discriminator', 'generator'])
    animator.fig.subplots_adjust(hspace=0.4)
    
    for epoch in range(1, num_epochs+1):
        timer = d2l.Timer()
        metric = d2l.Accumulator(3)
        for X, _ in data_iter:
            batch_size = X.shape[0]
            Z = torch.normal(0, 1, size=(batch_size, latent_dim, 1, 1))
            X, Z = X.to(device), Z.to(device)
            metric.add(d2l.update_D(X, Z, net_Disc, net_Gen, loss, trainer_Disc),
                       d2l.update_G(Z, net_Disc, net_Gen, loss, trainer_Gen),
                       batch_size)
        
        Z = torch.normal(0, 1, size=(21, latent_dim, 1, 1), device=device)
        
        fake_X = net_Gen(Z).permute(0, 2, 3, 1) / 2 + 0.5
        imgs = torch.cat(
            [torch.cat([
                fake_X[i*7+j].cpu().detach() for j in range(7)], dim=1)
            for i in range(len(fake_X)//7)], dim=0)
        
        animator.axes[1].cla()
        animator.axes[1].imshow(imgs)
        
        loss_Disc, loss_Gen = metric[0] / metric[2], metric[1] / metric[2]
        animator.add(epoch, (loss_Disc, loss_Gen))
    print(f'Discriminator loss: {loss_Disc:.4f},\nGenerator loss:     {loss_Gen: 4f}, \n{metric[2] / timer.stop():2f} examples/sec on {str(device)}')
```

```python
n_Dim_Gen = 32

net_Generator = nn.Sequential(
    GeneratorBlock(in_channels=64, out_channels=n_Dim_Gen*4, strides=1, padding=0), # Out: [64*4, 2, 2]
    GeneratorBlock(in_channels=n_Dim_Gen*4, out_channels=n_Dim_Gen*2),#*4),         # Out: [64*2, 4, 4]
    GeneratorBlock(in_channels=n_Dim_Gen*2, out_channels=n_Dim_Gen),#*2),           # Out: [64, 8, 8]
    #GeneratorBlock(in_channels=n_Dim_Gen*2, out_channels=n_Dim_Gen),                #Out: [64, 16, 16] 
    nn.ConvTranspose2d(in_channels=n_Dim_Gen, out_channels=3,
                       kernel_size=4, stride=2, padding=1, bias=False),             # Out: [3, 32, 32]
    nn.Tanh()   
)
```

```python
n_Dim_Disc = 32

net_Discriminator = nn.Sequential(
    DiscriminatorBlock(in_channels=3,          out_channels=n_Dim_Disc),           # Out: [64, 16, 16]
    DiscriminatorBlock(in_channels=n_Dim_Disc, out_channels=n_Dim_Disc*2),         # Out: [64*2, 8, 8]
    DiscriminatorBlock(in_channels=n_Dim_Disc*2, out_channels=n_Dim_Disc*4),       # Out: [64*4, 4, 4]
    #DiscriminatorBlock(in_channels=n_Dim_Disc*4, out_channels=n_Dim_Disc*8),       # Out: [64*8, 2, 2]
    nn.Conv2d(in_channels=n_Dim_Disc*4, out_channels=1, kernel_size=4, bias=False) # Out: [1, 1, 1]
)
```

```python
net_Generator = net_Generator.to(device)
net_Discriminator = net_Discriminator.to(device)

x = torch.zeros((1, 64, 1, 1)).to(device)
print(net_Generator(x).shape)

x = torch.zeros((1, 3, 32, 32)).to(device)
print(net_Discriminator(x).shape)
```

```python
latent_dim, lr, num_epochs = 64, 0.005, 20
train(net_Discriminator, net_Generator, data_iter, num_epochs, lr, latent_dim)
```
